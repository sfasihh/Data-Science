```
# PART B: Linear Regression for 100 Iterations

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
 
# Initialize Values
m = -0.5     # Slope
b = 0        # Y-intercept
l = 0.0001   # Learning rate
 
print('Slope: ' +str(m), 'Y-intercept: ' +str(b), 'Learning rate: ' +str(l))

url = 'https://raw.githubusercontent.com/tofighi/MachineLearning/master/datasets/student_marks.csv'   # Dataset
df1 = pd.read_csv(url)    # /Importing/Reading dataset


x = df1.iloc[:, 0]        # Initialize x to 'midterm marks'
y = df1.iloc[:, 1]        # Initialize y to 'final marks'

# Initial Regression Line
y1 = m*x + b

# Building Model for 100 Iterations

i = 100    # Iterations for Gradient Descent
n = float(len(x))
cost_update = []
m_update = []
b_update = []

# Finding Updated m, b and cost
for i in range(i):
      y_predicted = m * x + b                                      # Predicted Y
      cost = (1/n) * sum([val**2 for val in (y-y_predicted)])      # Defining Cost Function
      md = -(2/n)*sum(x*(y-y_predicted))                           # Defining Partial Derivatives
      bd = -(2/n)*sum(y-y_predicted)                               # Defining Partial Derivatives
      m = m - l * md                                               # Updating Slope using Learning Rate
      b = b - l * bd                                               # Updating Y-intercept using Learning Rate
      m_update.append(m)                                           # Appending updated m values into m_update
      b_update.append(b)                                           # Appending updated b values into b_update
      cost_update.append(cost)                                     # Appending updated cost values into cost_update

# Plotting Regression and Data-Set
y2 = m*x + b
plt.title('Midterm vs Final Marks')
plt.xlabel('Midterm Marks')
plt.ylabel('Final Marks')
plt.scatter(x, y)             # data set
plt.plot(x, y2, color='red')  # regression line
plt.show()

# Plotting Cost Function
for i in cost_update: 
    #print(i)
    plt.title('Cost Function')
    plt.xlabel('Iterations')
    plt.ylabel('Error')
    plt.plot(cost_update)   
```
